```python
import re
```

```python
with open('../../data/Day01.csv') as file:
    data = [line.strip() for line in file.readlines()]
    

print(data)
```

```python
spelled_out_digits = {'one' : 'o1e',
                      'two' : 't2o',
                      'three' : 't3e',
                      'four' : 'f4r',
                      'five' : 'f5e',
                      'six' : 's6x',
                      'seven' : 's7n',
                      'eight' : 'e8t',
                      'nine' : 'n9e'}

data_conv = data

for number in spelled_out_digits:
    data_conv = [line.replace(number, spelled_out_digits[number]) 
                 for line in data_conv]
    
print(data_conv)
```

```python
data_num = [re.findall(r'\d', line) for line in data_conv]
data_num
```

```python
data_dig = []
for line in data_num:
    data_dig.append(line[0] + line[-1])

data_dig = [int(line) for line in data_dig]
print(data_dig)
```

```python
results_day1 = sum(data_dig)
print(results_day1)
```
